```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import os 
import matplotlib
matplotlib.use('TkAgg')
```

```python
os.chdir("C:/Users/adiiz/Downloads/housing_price_dataset.csv")
data = pd.read_csv("housing_price_dataset.csv")
data.info()
data.describe()  #min price is negative
```

```python
#getting rid of all negative values
data = data[data['Price'] > 0] 
data = data.reset_index(drop=True)
data.info()
data_n= data['Neighborhood']
```

```python
# Convert categorical variable 'Neighborhood' to dummy variables
data = pd.get_dummies(data, columns=['Neighborhood'], drop_first=True)
data.head()
```

```python
X= data[['SquareFeet', 'Bedrooms', 'Bathrooms', 'YearBuilt', 'Neighborhood_Suburb', 'Neighborhood_Urban']]
y= data['Price']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

# Predict using the model
y_pred = regr.predict(X_test)

# Calculate accuracy
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
#graph for square feet vs price

plt.title("Price based on Square Feet")
plt.scatter(data['SquareFeet'], y)
plt.show()
```

```python
#graph for year built vs price

plt.title("Price based on Year Built")
plt.scatter(data['YearBuilt'], y)
plt.xlabel('year built')
plt.ylabel('price')
plt.show()
```

```python
#bar graph for neighborhood x price

plt.title("Price based on Neighborhood")
plt.bar(data_n, y) 
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.show()
```
